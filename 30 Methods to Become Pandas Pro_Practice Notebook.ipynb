### 3-4 Creating a DataFrame

`pd.DataFrame()`

```python
# From a list of lists

data = [[1, 2, "A"],
       [3, 4, "B"]]

df = pd.DataFrame(data, 
                 columns=["col1", "col2", 'col3'])

df
```

```python
# From a Dictionary

data = {'col1': [1, 2],
       'col2': [3, 4],
       'col3': ['A', 'B']}

df = pd.DataFrame(data=data)
df
```
